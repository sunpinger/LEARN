```python
import string
```

```python
def count():
    f = open('big.txt')
    word_freq = {}
    punct = string.punctuation
    for line in f:
        for c in punct:
            line = line.replace(c, "")
        words = line.split()
        for word in words:
            word = word.lower()
            if word in word_freq:
                word_freq[word] += 1
            else:
                word_freq[word] = 1
    for x,y in word_freq.items():
        print(x,y)
    # if len(args) != 0:
    #     print(f"Times: {word_freq[args[0]]}")
```

```python
count()
```

```python
def edits1(word): 
    # All edits that are one edit away from `word`.
    length = len(word)
    alphabet  = 'abcdefghijklmnopqrstuvwxyz'
    h1 = [word[:i]+word[i+1:] for i in range(length)]
    h2 = [word[:i]+j+word[i:] for i in range(length+1) for j in alphabet]
    h3 = [word[:i]+j+word[i+1:] for i in range(length) for j in alphabet]
    h4 = [word[:i]+word[i+1]+word[i]+word[i+2:] for i in range(length-1)]
    return set(h1+h2+h3+h4)

def edits2(word):
    # All edits that are one edit away from `word`.
    letters    = 'abcdefghijklmnopqrstuvwxyz'
    splits     = [(word[:i], word[i:])    for i in range(len(word) + 1)]
    deletes    = [L + R[1:]               for L, R in splits if R]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R)>1]
    replaces   = [L + c + R[1:]           for L, R in splits if R for c in letters]
    inserts    = [L + c + R               for L, R in splits for c in letters]
    return set(deletes + transposes + replaces + inserts)
```

```python
edits1('set')
```

```python
timeit edits1("set")
```

```python
timeit edits2("set")
```
